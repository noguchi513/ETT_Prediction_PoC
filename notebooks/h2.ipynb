```python
# Torch確認（Colab標準）
import torch
import torch.nn as nn

print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
```

```python
# ============================================================
# ETTh2 PatchTST + RevIN 改良完全版
# （baseline維持 / ジャンプ抑制 / 振幅改善）
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# ---------------------------
# 再現性
# ---------------------------
torch.manual_seed(42)
np.random.seed(42)

device = "cuda" if torch.cuda.is_available() else "cpu"

# ============================================================
# データ読み込み
# ============================================================
df = pd.read_csv("../data/ETTh2.csv")
df["date"] = pd.to_datetime(df["date"])
df = df.set_index("date")

target_col = "OT"
target_idx = df.columns.get_loc(target_col)

# ---------------------------
# Split
# ---------------------------
n = len(df)
train_end = int(n * 0.7)
val_end   = int(n * 0.85)

train_df = df.iloc[:train_end].copy()
val_df   = df.iloc[train_end:val_end].copy()
test_df  = df.iloc[val_end:].copy()

# ---------------------------
# StandardScaler
# ---------------------------
scaler = StandardScaler()

train_df.loc[:] = scaler.fit_transform(train_df)
val_df.loc[:]   = scaler.transform(val_df)
test_df.loc[:]  = scaler.transform(test_df)

# ============================================================
# RevIN
# ============================================================
class RevIN(nn.Module):
    def __init__(self, num_features, eps=1e-5):
        super().__init__()
        self.eps = eps
        self.gamma = nn.Parameter(torch.ones(1, 1, num_features))
        self.beta  = nn.Parameter(torch.zeros(1, 1, num_features))

    def forward(self, x, mode):

        if mode == "norm":
            self.mean = x.mean(1, keepdim=True)
            self.std  = x.std(1, keepdim=True)

            x = (x - self.mean) / (self.std + self.eps)
            return x * self.gamma + self.beta

        elif mode == "denorm":
            x = (x - self.beta) / self.gamma
            return x * (self.std + self.eps) + self.mean

# ============================================================
# Dataset
# ============================================================
INPUT_LEN = 168
PRED_LEN  = 168

class TimeSeriesDataset(Dataset):
    def __init__(self, data):
        self.data = data.values.astype(np.float32)

    def __len__(self):
        return len(self.data) - INPUT_LEN - PRED_LEN

    def __getitem__(self, idx):

        seq = self.data[idx : idx + INPUT_LEN + PRED_LEN]

        x = seq[:INPUT_LEN]
        y = seq[INPUT_LEN:, target_idx]

        return torch.tensor(x), torch.tensor(y)

train_ds = TimeSeriesDataset(train_df)
val_ds   = TimeSeriesDataset(val_df)
test_ds  = TimeSeriesDataset(test_df)

train_loader = DataLoader(train_ds, 64, shuffle=False)
val_loader   = DataLoader(val_ds, 64)
test_loader  = DataLoader(test_ds, 64)

# ============================================================
# Patch Embedding
# ============================================================
class PatchEmbedding(nn.Module):
    def __init__(self, in_dim, patch_len, d_model):
        super().__init__()
        self.patch_len = patch_len
        self.proj = nn.Linear(in_dim * patch_len, d_model)

    def forward(self, x):

        B, L, C = x.shape
        num_patch = L // self.patch_len

        x = x[:, : num_patch * self.patch_len]

        x = x.reshape(
            B,
            num_patch,
            self.patch_len * C
        )

        return self.proj(x)

# ============================================================
# PatchTST 改良版
# ============================================================
class PatchTST(nn.Module):
    def __init__(
        self,
        in_dim,
        patch_len=24,
        d_model=512,
        nhead=8,
        layers=4
    ):
        super().__init__()

        self.revin = RevIN(in_dim)

        self.patch = PatchEmbedding(
            in_dim,
            patch_len,
            d_model
        )

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            batch_first=True
        )

        self.encoder = nn.TransformerEncoder(
            encoder_layer,
            num_layers=layers
        )

        self.head = nn.Linear(
            d_model,
            PRED_LEN
        )

    def forward(self, x):

        # -----------------------
        # 直前値保存（baseline維持）
        # -----------------------
        last_value = x[:, -1, target_idx]

        # -----------------------
        # RevIN 正規化（入力安定化）
        # -----------------------
        x = self.revin(x, "norm")

        # -----------------------
        # Patch + Transformer
        # -----------------------
        x = self.patch(x)
        x = self.encoder(x)

        # -----------------------
        # mean pooling廃止
        # -----------------------
        x = x[:, -1]

        # -----------------------
        # 予測
        # -----------------------
        x = self.head(x)

        # -----------------------
        # Residual接続
        # -----------------------
        x = x + last_value.unsqueeze(-1)

        return x

# ============================================================
# モデル
# ============================================================
model = PatchTST(
    in_dim=len(df.columns)
).to(device)

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=1e-4
)

loss_fn = nn.MSELoss()

# ============================================================
# 学習
# ============================================================
EPOCHS = 70

for epoch in range(EPOCHS):

    model.train()
    total_loss = 0

    for x, y in tqdm(
        train_loader,
        desc=f"Epoch {epoch+1}"
    ):
        x = x.to(device)
        y = y.to(device)

        optimizer.zero_grad()

        out = model(x)

        loss = loss_fn(out, y)
        loss.backward()

        optimizer.step()

        total_loss += loss.item() * x.size(0)

    print(
        f"Epoch {epoch+1} Loss: "
        f"{total_loss / len(train_loader.dataset):.4f}"
    )

# ============================================================
# 評価
# ============================================================
def evaluate(loader):

    model.eval()
    preds, trues = [], []

    with torch.no_grad():
        for x, y in loader:

            x = x.to(device)
            out = model(x)

            preds.append(out.cpu().numpy())
            trues.append(y.numpy())

    preds = np.concatenate(preds).flatten()
    trues = np.concatenate(trues).flatten()

    mae  = mean_absolute_error(trues, preds)
    rmse = mean_squared_error(trues, preds) ** 0.5

    return mae, rmse

mae, rmse = evaluate(test_loader)

print(f"\nTest MAE : {mae:.4f}")
print(f"Test RMSE: {rmse:.4f}")

# ============================================================
# 可視化（逆標準化）
# ============================================================
x, y = next(iter(test_loader))
x = x.to(device)

with torch.no_grad():
    pred = model(x[:1])

mean = scaler.mean_[target_idx]
std  = scaler.scale_[target_idx]

true_plot = y[0].numpy() * std + mean
pred_plot = pred[0].cpu().numpy() * std + mean

plt.figure(figsize=(12, 5))

plt.plot(true_plot, label="True")
plt.plot(pred_plot, label="Pred")

plt.legend()
plt.title("PatchTST + RevIN Improved Forecast")
plt.show()
```
